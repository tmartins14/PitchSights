```python
from pathlib import Path
import pandas as pd

p = Path("team_ratings_snapshots.parquet")  # your path

print(p, "size:", p.stat().st_size, "bytes")

# Inspect magic bytes: Parquet must start and end with b'PAR1'
with open(p, "rb") as f:
    head = f.read(4)
    f.seek(-4, 2)
    tail = f.read(4)
print("head:", head, "tail:", tail)

# Try reading as parquet, else as csv to see what's inside
try:
    dfp = pd.read_parquet(p)
    print("Parquet columns:", dfp.columns.tolist()[:10])
except Exception as e:
    print("Parquet read failed ->", e)
    try:
        dfc = pd.read_csv(p)
        print("Looks like CSV. Columns:", dfc.columns.tolist()[:10])
    except Exception as e2:
        print("CSV read also failed ->", e2)
```
